```python
# import pyspark - importando
import pyspark
# get access to spark - Acessando a biblioteca 
spark = pyspark.sql.SparkSession(pyspark.SparkContext())
spark
```

```python
# Convert Excel files to csvs - Covertendo de xlsx para csv (para trabalhar de uma forma mais f√°cil)
import pandas as pd #importando a biblioteca pandas e subtituindo seu nome por 'pd'

def excel_to_csv(name):
  pd.read_excel(name + '.xlsx').to_csv(name + '.csv', sep=',', index=False)

excel_to_csv('Athletes')
excel_to_csv('Coaches')
excel_to_csv('EntriesGender')
excel_to_csv('Medals')
excel_to_csv('Teams')
```

```python
# Read in Pandas dataframes
athletes = pd.read_csv('Athletes.csv')
coaches = pd.read_csv('Coaches.csv')
entries_gender = pd.read_csv('EntriesGender.csv')
medals = pd.read_csv('Medals.csv')
teams = pd.read_csv('Teams.csv')
```

```python
# sort medals by total- Rank organizado por ordem de total de medalhas 
medals.sort_values('Rank by Total', inplace=True)
medals.head(5)
```

```python
# make bar chart of medals won by each country showing gold, silver and bronze- Cria√ß√£o de um gr√°fico de barra para avaliar os pa√≠ses com maior desempenho 
# De Acordo com as medalhas ganhas =)
import plotly.express as px
fig = px.bar(medals, x='Team/NOC', y=['Gold', 'Silver', 'Bronze'], color_discrete_sequence =['gold', 'silver', 'black'],title='Number of medals won by each country')
fig.show()
```

```python
# Add coaches to pyspark sql and make most basic query
spark.read.csv('Coaches.csv', header=True, inferSchema=True).createOrReplaceTempView('coaches')
spark.sql('select * from coaches').show(5)
```

```python
# Obtain the count of coaches from each NOC (country) - Querry para obter a contagem atrav√©s da linguaguem SQL 
# Fazendo contagem dos n√∫meros de treinadores por pa√≠ses
pd_coaches_query = spark.sql('select NOC, count(NOC) as count from Coaches group by NOC').toPandas()
pd_coaches_query.head()
```

```python
spark.read.csv('Teams.csv', header=True, inferSchema=True).createOrReplaceTempView('Teams')
spark.sql('select * from Teams').show(5)
```

```python
# Simple bar chart of coaches from each country - Gr√°fico de barra simples com a quantidade de treinador por pa√≠ses
# Obtemos um mapa de calor de acordo com a quantidade de treinadores/pa√≠ses
fig = px.bar(pd_coaches_query, x='NOC', y='count', color='count',title='Number of coaches from each country')
fig.show()
```

```python
# Show head of athletes - Mostrando tabela de atletas
athletes.head()
```

```python
# Add athletes to pyspark sql - Adicionando ao banco de dados para fazer uma consulta com SQL
spark.read.csv('Athletes.csv', header=True, inferSchema=True).createOrReplaceTempView('athletes')
```

```python
# Get count of each unique combination of NOC and Discipline
# Consultando atrav√©s de uma linguaguem SQP 
pd_athletes_query = spark.sql('''select first(NOC) as NOC, first(Discipline) as Discipline, 
                                 count(*) Count from athletes group by NOC, Discipline order 
                                 by NOC, Discipline''').toPandas()
pd_athletes_query.head(10)
```

```python
# Bar chart of number of athletes split by discipline -  Um Gr√°fico de barra com n√∫mero de atlestas por modalidade
fig = px.bar(pd_athletes_query, x='NOC', y='Count', color='Discipline', title='Number of athletes from each country per discipline')
fig.show()
```

```python
# Show amount of each discipline - Para contar os valores de atletas por casda modalidade 
athletes['Discipline'].value_counts().head(10)
```

```python
# Only keep rows with discipline in top 3
pd_athletes_query = pd_athletes_query.loc[pd_athletes_query['Discipline'].isin(['Athletics', 'Swimming', 'Football'])]
pd_athletes_query.head(5)
```

```python
# Bar chart of number of athletes split by discipline
fig = px.bar(pd_athletes_query, x='NOC', y='Count', color='Discipline', title='Number of athletes from each country per discipline')
fig.show()
```
